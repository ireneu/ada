```python
import folium
import json
import numpy as np
import pandas as pd
import random
```

### First we will create the necessary map

```python
topo_json_path = 'ch-cantons.topojson.json'
    
cantons = [canton['id'] for canton in json.load(open(topo_json_path))['objects']['cantons']['geometries']]
df = pd.DataFrame([[canton, random.randint(1, 50)] for canton in cantons], columns=['Canton', 'Grants'])
df
```

```python
canton_map = folium.Map(location=[46.8, 8.2],
                        zoom_start=8)

canton_map.choropleth(geo_path=topo_json_path,
                      data=df,
                      columns=['Canton', 'Grants'],
                      key_on='feature.id',
                      fill_color='BuPu',
                      legend_name='Grants',
                      topojson='objects.cantons')

canton_map
```

```python
snsf_data = pd.read_csv('P3_GrantExport.csv', sep=';')

# Drop rows with no University
snsf_data = snsf_data.dropna(axis=0, subset=['University'])
snsf_data = snsf_data[(snsf_data.University != 'Nicht zuteilbar - NA') & (snsf_data.University != 'NPO (Biblioth., Museen, Verwalt.) - NPO')]

snsf_data.sample(5)
```

```python
unis = snsf_data['University'].unique()
unis
```

```python
len(unis)
```

```python
import requests
```

```python
l = list()
for uni in unis:
    response = json.loads(requests.get('http://api.geonames.org/searchJSON?q="{0}"&username=ada4ever&maxrows=15&country=CH'.format(uni.split('-')[0])).text)
    if response.get('totalResultsCount') == 0 and len(uni.split('-')) > 1:
        response = json.loads(requests.get('http://api.geonames.org/searchJSON?q="{0}"&username=ada4ever&maxrows=15&country=CH'.format(uni.split('-')[1])).text)
    if response.get('totalResultsCount') != 0:
        l.append((uni, response['geonames'][0]['adminName1']))

l
```
